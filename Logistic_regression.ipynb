```python
from datasets import load_dataset

dataset = load_dataset("sst", "default")
```

```python
import numpy as np
# Assuming you want to convert the training split to a NumPy array
training_data = dataset["train"]
# Convert the features and labels to NumPy arrays
features = np.array(training_data["sentence"])
labels = np.array(training_data["label"])
tokens = np.array(training_data["tokens"])
# Initialize empty lists to store sentences, labels, and tokens
sentences = []
labels = []
result_array = []
# Iterate over training data and extract sentences and labels

for example in training_data:
    sentence = example["sentence"]
    label = example["label"]
    result_array.append((sentence , label))

result_array = np.array(result_array)
print(result_array)
```

```python
# Define sentiment categorization function
def categorize_sentiment(sentiment):
    if 0 <= sentiment <= 0.2:
        return 0
    elif 0.2 < sentiment <= 0.4:
        return 1
    elif 0.4 < sentiment <= 0.6:
        return 2
    elif 0.6 < sentiment <= 0.8:
        return 3
    elif 0.8 < sentiment <= 1.0:
        return 4
    else:
        raise ValueError("Sentiment value out of range (0-1)")


for i in range(len(result_array)):
    # Lowercase the phrase
    phrase = result_array[i, 0].lower()
    # Remove punctuation
    # phrase = phrase.translate(str.maketrans('', '', string.punctuation))
    # turn number to category
    sentiment_class = categorize_sentiment(float(result_array[i, 1]))
    result_array[i , 1] = sentiment_class
    result_array[i,0] = phrase
```
